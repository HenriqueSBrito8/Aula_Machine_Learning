```
################################################################
################### REGRESSÃO LINEAR SIMPLES ###################
################################################################

################################################################
################# MODELO 1 - TREINO E TESTE ####################
################################################################
```

```
# Criando x e y
# Cria o dataset. Criamos em formato de array. O reshape cria em duas dimensões (uma coluna e várias linhas)
x = np.array(df_menor['GrLivArea']).reshape((-1, 1)) # variável independente x
y = np.array(df_menor['SalePrice']) # variável dependente y
```

```
# Separando os dados de Treino e Teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Print do número de linhas e o número de colunas
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
# Para visualizar os array
# display(x_train)
# display(x_test)
# display(y_train)
# display(y_test)
```

```
# Cria o modelo de regressão linear e aplica o modelo (fit)
modelo_1 = LinearRegression()

# Treinando o Modelo
modelo_1.fit(x_train,y_train)

# ... ou assim em uma linha
# modelo_1 = LinearRegression().fit(x_train,y_train)
# modelo_1
```

```
# Intercepto = representa o valor de "a" ou "b0" da equação da regressão linear
# O valor encontrado é a previsão do modelo quando o valor de X é zero.
print('Intercepto (a ou b0):', modelo_1.intercept_)

# O coeficiente = representa o “peso” da variável no modelos neste caso, na equação identificado como b1
# Significa desse valor é previsão da adição de cada valor do coeficiente quando o valor de x é aumentado em um.
# Ou seja, quando x aumenta em 1 é adicionado o valor de b1 no coeficiente da variável independente sobre o valor a ser previsto.
print('Coeficiente (b1):', modelo_1.coef_)
```

```
############## Importante
############## Outra forma de fazer a regressão para verificar a significância das variáveis no modelo
############## Olhar para a coluna "P>|t|" se o valor for < 0.05, ela é signficativa, permanece no modelo. Caso contrário, é retirada
```

```
# Adicione uma constante ao x_train para representar o intercepto
x_train_with_const = sm.add_constant(x_train)

# Crie o modelo de regressão com statsmodels
modelo_stats = sm.OLS(y_train, x_train_with_const).fit()

# Obtenha um resumo das estatísticas do modelo
summary = modelo_stats.summary()

# Exiba o resumo com os valores-p
print(summary)

# SalePrice (y) = 1.333e+04 (a) + 110.2643 (b) * GrLivArea (x) - GrLivArea ela é significativa (P > |T| < 0.05)
```

```
############## Importante
############## Outra forma de fazer a regressão para verificar a significância das variáveis no modelo
```

```
# Equação Geral >> y = a + bx
# Equação para nosso modelo >> SalePrice = 13330.3 + 110.3 * GrLivArea
```

```
# Fazendo a previsão dos preços das casas, e o print da casa de indice 10
y_pred = modelo_1.predict(x_test)
y_pred[10]
```

```
# Criando um dataframe com os valores do teste (reais) e da previsão (preditos)
df1=pd.DataFrame({'Real': y_test, 'Predicted':y_pred})

# Formatando os valores preditos
df1['Predicted'] = round(df1['Predicted'],3)

# Criando uma razão entre o valor predito e o real para saber o quanto está próximo ou não do real
df1['Razao'] = round(df1['Predicted']/df1['Real'],3)

print(df1.shape)
df1.head(15)
```

```
# Métricas de avaliação

# Erro Absoluto Médio (MAE): mede as distâncias entre valores preditos e observados.
# Essa métrica não é muito influenciada pelos outliers. Quanto menor seu valor, melhor
MAE_M1 = metrics.mean_absolute_error(y_test, y_pred)
print('MAE_M1:', metrics.mean_absolute_error(y_test, y_pred))

# Erro Quadrático Médio (MSE): indicada para problemas nos quais grandes erros não são tolerados.
# Mais difícil de interpretar, pois eleva a unidade ao quadrado. Quanto menor seu valor, melhor
MSE_M1 = metrics.mean_squared_error(y_test, y_pred)
print('MSE_M1:', metrics.mean_squared_error(y_test, y_pred))

# Raiz do Erro Quadrático Médio (RMSE): tem a mesma unidade dos valores observados e preditos.
# Seu resultado é uma medida do desvio médio (erro) entre observado e predito
RMSE_M1 = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('RMSE_M1:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# R2: calcula qual a porcentagem da variância que pode ser explicada pelo modelo de regressão.
# Informa o quão “próximo” as medidas reais estão do nosso modelo. Pode variar entre 0 e 1
R2_SCORE_M1 = metrics.r2_score(y_test, y_pred)
print('R2_SCORE_M1:', metrics.r2_score(y_test, y_pred))
```
