```
# Objetivo: prever o valor de venda (SalePrice) por alguma variável
# Por meio da regressão => y = a + bx + e >>> SalePrice = a + b*GrLivArea + e

# Iremos escolher a GrLivArea como variável explicativa
```

```
# Verificando visualmente se há alguma associação entre SalePrice e GrLivArea

# Gráfico de dispersão entre GrLivArea e SalePrice
plt.title('Relação entre a variavel X e Y')
plt.scatter(x=df['GrLivArea'], y=df['SalePrice'])
plt.show()
```

```
# A variável SalePrice será variável DEPENDENTE ou RESPOSTA para a regressão linear simples
# A variável GrLivArea será variável INDEPENDENTE ou EXPLICATIVA para a regressão linear simples
```

```
################################################################
################### REGRESSÃO LINEAR SIMPLES ###################
################################################################

################################################################
######## MODELO 1.0 - SEM TRANSFORMAÇÃO NAS VARIÁVEIS ##########
################################################################
```

```
# Criando x e y
# Cria o dataset. Criamos em formato de array. O reshape cria em duas dimensões (uma coluna e várias linhas)
x = np.array(df['GrLivArea']).reshape((-1, 1)) # variável independente x
y = np.array(df['SalePrice']) # variável dependente y
```

```
# Verificando o x
x
```

```
# Verificando o y
y
```

```
# Cria o modelo de regressão linear e aplica o modelo (fit)
modelo = LinearRegression()
modelo.fit(x,y)

# ... ou assim em uma linha
# modelo = LinearRegression().fit(x,y)
# modelo
```

```
# Avalia o modelo
print('coeficiente de determinação:', round(modelo.score(x, y),4))

# Intercept
print('intercept:',(modelo.intercept_))

# Slope
print('slope:', modelo.coef_)
```

```
# y = a + bx >>> a e b são coeficientes
# y = 18569.025856 + 107.13035897 * GrLivArea
```

```
# Define o tema do gráfico
sns.set_theme()

# Define o tamanho do gráfico
plt.figure(figsize = (20, 10))

# Cria o gráfico de dispersão com a Reta de regressão
sns.regplot(x = 'GrLivArea', y = 'SalePrice', data = df)
```

```
# Outra forma de fazer o gráfico anterior
# Define o tamanho do gráfico
plt.figure(figsize=(20, 10))

# Cria o gráfico de dispersão
plt.scatter(df['GrLivArea'], df['SalePrice'], alpha=0.5, label='Dados')

# Ajusta a reta de regressão
slope, intercept = np.polyfit(df['GrLivArea'], df['SalePrice'], 1)
plt.plot(df['GrLivArea'], slope * df['GrLivArea'] + intercept, color='red', label='Reta de Regressão')

# Configurações adicionais
plt.title('Gráfico de Dispersão com Reta de Regressão')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.legend()

# Exibe o gráfico
plt.show()
```

```
# Outra forma - regressão linear simples
# Olhar a significância dos parâmetros ou coeficientes

# ols_modelo = ols("y ~ x", data=df)

ols_modelo = ols("y ~ x", data=df)
modelo = ols_modelo.fit()
print(modelo.summary())
```

### Modelo final

###### Como o coeficiente linear e angular são significativos (P > |t| < 0.05), temos:
###### - SalePrice = intercept: 18570 + 107.1304 * GrLivArea
###### - R2 = 0.502
###### - Possivelmente outras variáveis podem contribuir para a definição do preço das casas, além da área da sala de estar
